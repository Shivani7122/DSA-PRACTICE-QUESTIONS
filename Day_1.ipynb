```python
class Solution:
    def twoSum(self, nums, target):
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i,j]
                else:
                    pass
        return []
obj = Solution()
num = []
n = int(input("Enter the length of array: "))
for i in range(n):
  num.append(int(input("Enter the numbers: ")))
print(num)
target = int(input("Enter the target Sum: "))
obj.twoSum(num,target)
```

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        if x < 0:
            return False
        s = str(x)
        return s == s[::-1]
obj = Solution()
x = int(input("Enter the number: "))
obj.isPalindrome(x)

```

```python
class Solution:
    def longestCommonPrefix(self, strs):
        if not strs:
            return ""
        prefix = strs[0]
        for string in strs[1:]:
            while not string.startswith(prefix):
                prefix = prefix[:-1]
                if not prefix:
                    return ""
        return prefix

obj = Solution()
strs = ["windy","wind","windtalker"]
obj.longestCommonPrefix(strs)
```

```python
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        num1 = int(a,2)
        num2 = int(b,2)
        total = num1+num2

        return bin(total)[2:]
obj = Solution()
obj.addBinary("1010","1011")
```
